**AI and Advertisement:** Develop AI-powered programmatic advertising by using cookies and collecting user data.

```python
import numpy as np
from sklearn.ensemble import RandomForestClassifier
```

```python
# Training data format:
# [hour_of_day, location_code] → click label (1 = click, 0 = no click)

training_data = np.array([
    [10, 1],
    [14, 2],
    [18, 1],
    [20, 2],
    [12, 1],
])

labels = np.array([1, 0, 1, 0, 1])
```

```python
model = RandomForestClassifier()
model.fit(training_data, labels)
```

```python
ads_pool = [
    {"ad_text": "Buy Shoes - 50% Off!", "tag": 1},
    {"ad_text": "Try New Coffee!", "tag": 2},
    {"ad_text": "Eat and Pay 50%!", "tag": 3}
]
```

```python
def get_ad(hour: int, location: str):
    loc = location.lower()

    if loc == "india":
        loc_code = 1
    elif loc in ["usa", "us"]:
        loc_code = 2
    else:
        loc_code = 3  # Other countries

    features = [hour, loc_code]

    click_prob = model.predict_proba([features])[0][1]

    if click_prob > 0.7:
        return ads_pool[0]  # High chance
    elif click_prob > 0.4:
        return ads_pool[1]  # Medium chance
    else:
        return ads_pool[2]  # Low chance
```

```python
hour = 8
location = "INDIA"

recommended_ad = get_ad(hour, location)

print(f"Recommended Ad for {location} at {hour}:")
print(recommended_ad["ad_text"])
```

```python
hour = 20
location = "USA"

result = get_ad(hour, location)

print(f"Recommended Ad for {location} at {hour}:")
print(result["ad_text"])
```
