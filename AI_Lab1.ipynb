**Marketer to Machine:** Develop a ML model for Smart email Compose. Smart email compose finishes sentences for you by predicting what word or words user will type next.

```python
# Importing Libraries
import random
from collections import defaultdict
```

```python
def train_model(sentences):
    word_dict = defaultdict(list)

    for sentence in sentences:
        words = sentence.split()

        for i in range(len(words) - 1):
            word_dict[words[i]].append(words[i + 1])

    return word_dict
```

```python
def predict_next_word(word_dict, text):
    words = text.split()
    last_word = words[-1] if words else ""

    return random.choice(word_dict.get(last_word, ["..."]))
```

```python
# Sample training data
sentences = [
    "Hello how are you",
    "I am doing great",
    "Nice to meet you",
    "Hope you are having a good day",
    "Looking forward to our meeting"
]

word_dict = train_model(sentences)
```

```python
# Get user input and predict next word
user_input = input("Type a sentence: ")
next_word = predict_next_word(word_dict, user_input)
print(f"Predicted next word: {next_word}")
```
